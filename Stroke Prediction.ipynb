```python
x = df.iloc[:,:-1]   # In x taking all columns except the last column
x
```

```python
y = df.iloc[:,-1]   # In y taking only last column
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)   # Splitting x and y into train and test data with test_size=0.2 i.e 20% data will be taken as test data and 80% will be taken as train data
```

```python
# Importing required modules
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
```

```python
# Converting Following columns into numbers using OneHotEncoder
ohe = OneHotEncoder()
ohe.fit(x[['gender','ever_married','work_type','Residence_type','smoking_status']])
```

```python
ohe.categories_   # Checking categories in ohe
```

```python
# Importing required modules
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
# First it will use ohe to convert given columns and the reamning columns are passed with remainder='passthrough'
colum_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),
                                      ['gender','ever_married','work_type','Residence_type','smoking_status']),
                                     remainder='passthrough')
```

```python
# Taing DecisionTreeClassifier into clf veriable
clf = DecisionTreeClassifier(random_state=0)
```

```python
# First pipeline runs colum_trans and pass it to clf and all this will be stored into pipe
pipe = make_pipeline(colum_trans,clf)
```

```python
# Fitting x_train and y_train into pipe
pipe.fit(x_train,y_train)
```

```python
# Predicting output on x_test
y_pred = pipe.predict(x_test)
y_pred
```

```python
# Checking accurcy of model
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy: {:.2f}%".format(accuracy))
```

```python
# Finding best accuracy

for i in range(1000):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
  clf = DecisionTreeClassifier(random_state=0)
  pipe = make_pipeline(colum_trans,clf)
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test)
  print(accuracy_score(y_test,y_pred),i)
```

```python
# Checking the random_state where we get higest accuracy

score = []
for i in range(1000):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
  clf = DecisionTreeClassifier(random_state=0)
  pipe = make_pipeline(colum_trans,clf)
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test)
  score.append(accuracy_score(y_test,y_pred))
```

```python
score[np.argmax(score)]
```

```python
# Refitting the model with new random_state

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=np.argmax(score))
clf = DecisionTreeClassifier(random_state=0)
pipe = make_pipeline(colum_trans,clf)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
accuracy_score(y_test,y_pred)
```
